```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

 ## Trie et préparation des données

```python
data = {
    "real" : [],
    "user" : [],
    "sys" : []
}

nbFile = []
for i in range(100):
    nbFile.append(30 * (i + 1))

    
with open("data_tf.txt", "r") as f:
    count = 0
    lines = f.readlines()
    

    for line in lines:
        
        line = float(line.split(" ")[1].replace(",", "."))
        if count == 0:
            data["real"].append(line)
            count += 1
            
        elif count == 1:
            data["user"].append(line)
            count += 1
            
        elif count == 2:
            data["sys"].append(line)
            count = 0
            
print(len(nbFile))
print(len(data["sys"]))

data_tf = data
```

```python
# Fonction linéaire
def func(x, a, b, ):
    return a*x + b
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

#Scatter point
color = ["#d90429", "#73ba9b", "#339cff"]
label = ["real", "user", "sys"]
for i in range(len(data)):
    
    ax.scatter(data[label[i]], nbFile, color = color[i], label=label[i])
    
    # Courbe de tendance
    popt, pcov = curve_fit(func, data[label[i]], nbFile)
    ax.plot(data[label[i]], np.polyval(popt, data[label[i]]), '-', label = "Droite de tendance : " + label[i], color = "#212121")

ax.legend()
ax.set(title="Observation des temps de calcul de la commande TF sur des différentes quantités de fichiers", xlabel = "Temps (en seconde)", ylabel = "Nombre de fichier")

fig.savefig("tf.jpg")
```
